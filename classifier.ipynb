```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications.resnet import preprocess_input
from functools import partial
from PIL import Image
import matplotlib.pyplot as plt
import copy
import cv2
import csv
```

```python
#set seed so results are reproducible
seed = 1
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# Model / data parameters
num_classes = 6
input_shape = (64, 64, 3)

# load labeled train data
x_t = []
labeled_y_tr = []

for l in range(0,6):
  for i in os.listdir(dir+"lableed/"+str(l)):
    f = os.path.join(dir + "lableed/"+str(l), i)
    im = cv2.imread(f)
    x_t.append(im)
    labeled_y_tr.append(l)

labeled_x_train = np.array(x_t)



# load unlabeled train data
x_t_unlabeled = []
for i in os.listdir(dir+"unlabeled"):
  f = os.path.join(dir+"unlabeled", i)
  im = cv2.imread(f)
  x_t_unlabeled.append(im)

unlabeled_x_train = np.array(x_t_unlabeled)



#load testing data
x_te_unlabeled = []
for i in os.listdir(dir+"test"):
  f = os.path.join(dir+"test", i)
  im = cv2.imread(f)
  x_te_unlabeled.append(im)


x_test = np.array(x_te_unlabeled)


#Data preprocessing
unlabeled_x_train_scaled = preprocess_input(unlabeled_x_train)
labeled_x_train_scaled = preprocess_input(labeled_x_train)
x_test_scaled = preprocess_input(x_test)

#one hot encode labels
labeled_y_tr = np.array(labeled_y_tr)
labeled_y_train = keras.utils.to_categorical(labeled_y_tr, num_classes)


#training data shape
print("labeled training data shapes")
print(labeled_x_train_scaled.shape)
print(labeled_y_train.shape)

print("unlabeled data shapes")
print(unlabeled_x_train_scaled.shape)

print("test data shapes")
print(x_test_scaled.shape)

```

```python
class SemisupervisedModel(keras.Model):
    def __init__(self, model, entropy=False, lamda=0.1):
        super().__init__()

        self.model = model
        self.cce = tf.keras.losses.CategoricalCrossentropy()

        self.entropy = entropy
        self.lamda = lamda

    def compile(self, optimizer, **kwargs):
        super().compile(**kwargs)

        self.optimizer = optimizer
        self.accuracy_tracker = keras.metrics.Accuracy()
        self.loss_tracker = keras.metrics.Mean(name="loss")

    @property
    def metrics(self):
        return [self.loss_tracker, self.accuracy_tracker]
        # return [self.loss_tracker]

    def call(self, images, training=True):
        return self.model(images, training=training)

    def semisupervised_loss(self, labeled_y_true, labeled_y_pred,  unlabeled_y_pred):
        loss = self.cce(labeled_y_true, labeled_y_pred)
        if self.entropy:
            loss += -self.lamda * tf.reduce_mean(tf.reduce_sum(unlabeled_y_pred * tf.math.log(unlabeled_y_pred + 1e-12), axis=1))
        return loss

    def train_step(self, data):
        (unlabeled_images), (labeled_images, labeled_y_true) = data

        with tf.GradientTape() as tape:
            labeled_y_pred = self.call(labeled_images, training=True)
            unlabeled_y_pred = self.call(unlabeled_images, training=True)
            loss = self.semisupervised_loss(labeled_y_true, labeled_y_pred, unlabeled_y_pred)
        gradients = tape.gradient(loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(gradients, self.trainable_weights))
        self.loss_tracker.update_state(loss)

        #self.accuracy_tracker.update_state(tf.math.argmax(unlabeled_y_true, axis=1), tf.math.argmax(unlabeled_y_pred, axis=1))
        self.accuracy_tracker.update_state(tf.math.argmax(labeled_y_true, axis=1), tf.math.argmax(labeled_y_pred, axis=1))

        return {m.name: m.result() for m in self.metrics}

    def test_step(self, data):
        images, labels = data
        y_pred = self.call(images, training=False)

        loss = self.cce(labels, y_pred)
        self.loss_tracker.update_state(loss)

        self.accuracy_tracker.update_state(tf.math.argmax(labels, axis=1), tf.math.argmax(y_pred, axis=1))

        return {m.name: m.result() for m in self.metrics}
```

```python
BATCH_SIZE = 700
SHUFFLE_BUFFER_SIZE = 50

labeled_train_dataset = tf.data.Dataset.from_tensor_slices((labeled_x_train_scaled, labeled_y_train))
labeled_train_dataset = labeled_train_dataset.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
unlabeled_train_dataset = tf.data.Dataset.from_tensor_slices((unlabeled_x_train_scaled))
unlabeled_train_dataset = unlabeled_train_dataset.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
train_dataset = tf.data.Dataset.zip((unlabeled_train_dataset, labeled_train_dataset)).prefetch(buffer_size=tf.data.experimental.AUTOTUNE)

test_dataset = tf.data.Dataset.from_tensor_slices((x_test_scaled))
test_dataset = test_dataset.batch(BATCH_SIZE)
```

```python
base_model = tf.keras.applications.ResNet50(
    include_top=False,
    weights="imagenet",
    input_shape=(64,64,3),
    classes=6,
)

for layer in base_model.layers:  
    layer.trainable = False
#Open up final layers
for layer in base_model.layers[45:]:
   layer.trainable = True


semi_model = tf.keras.Sequential([
  base_model,
  layers.GlobalAveragePooling2D(),
  layers.BatchNormalization(),
  layers.Dense(50,activation='elu',kernel_regularizer=tf.keras.regularizers.l1(0.03),
                              activity_regularizer=tf.keras.regularizers.l2(0.015)),
  layers.BatchNormalization(),
  layers.Dense((6),activation='softmax')
])
```

```python
epochs = 10


semi_model.compile(optimizer=keras.optimizers.Adam(lr=0.0004),loss= 'categorical_crossentropy' ,metrics=["accuracy"])
#semi_model.compile(optimizer=keras.optimizers.Adam(lr=0.0004), metrics=["accuracy"])


X_TR = np.copy(labeled_x_train_scaled)
y_TR = np.copy(labeled_y_train)
X_TR_UNLABELED = np.copy(unlabeled_x_train)

iters = 6

new_labels = 1

semi_model.fit(X_TR, y_TR, epochs=epochs)

```

```python
yhat_unlabeled = semi_model.predict(X_TR_UNLABELED)
new_labels = 0

#extract high confidence prediction to add to labeled data
y_p = []
x_p = []
p_indices = []
for p in range(0,len(yhat_unlabeled)):
    if max(yhat_unlabeled[p]) >= 0.90:
          y_pseudo = np.zeros(num_classes)
          max_index = tf.argmax(yhat_unlabeled[p], axis=0)
          y_pseudo[int(max_index)] = 1
          y_p.append(y_pseudo)
          p_indices.append(p)
          x_p.append(X_TR_UNLABELED[p])
          new_labels += 1

    y_p = np.array(y_p)
    x_p = np.array(x_p)

    if len(y_p) > 0:
        X_TR = np.concatenate((X_TR, x_p), axis=0)
        y_TR = np.concatenate((y_TR,y_p),axis=0)
        X_TR_UNLABELED = np.delete(X_TR_UNLABELED,p_indices,axis=0)


semi_model.fit(X_TR, y_TR, epochs=epochs)
```
